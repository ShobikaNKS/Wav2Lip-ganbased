```python
import edge_tts
import asyncio
import os

async def generate_audio():
    text = """
    Namaste Mathangi! My name is Anika, and I’m here to guide you through managing your credit card dues.
    Your credit card bill shows an amount due of INR 5,053 which needs to be paid by 31st December 2024.
    Please make your payment through UPI or bank transfer. Thank you!
    """

    audio_file = "unique_audio_edge_tts.wav"


    tts = edge_tts.Communicate(text, "en-IN-NeerjaNeural")
    await tts.save(audio_file)

    print(f"Audio saved as {audio_file}")
    return audio_file


try:

    if not asyncio.get_running_loop():
        audio_file = asyncio.run(generate_audio())
    else:
        audio_file = await generate_audio()
except RuntimeError:
    audio_file = await generate_audio()


output_video = "lip_synced_unique_output.mp4"
input_image = "input1.jpeg"


if os.path.exists(audio_file):
    os.system(f'python inference.py --checkpoint_path checkpoints/wav2lip_gan.pth --face "{input_image}" --audio "{audio_file}" --outfile "{output_video}"')
    print(f"Lip-synced video generated: {output_video}")
else:
    print("Audio file not found. Lip-syncing aborted.")
```
